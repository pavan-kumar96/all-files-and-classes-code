```python
url = 'https://raw.githubusercontent.com/jpatokal/openflights/master/data/airports.dat'
import urllib.request
response = urllib.request.urlopen(url)
data = response.read()
fp = open('airport_data.csv', 'wb')
fp.write(data)
fp.close()
```
