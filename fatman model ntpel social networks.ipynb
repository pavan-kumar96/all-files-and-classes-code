```python
import networkx as nx
import matplotlib.pyplot as  plt
import random
```

```python
def create_graph():
    G = nx.Graph()
    G.add_nodes_from(range(1,101))
    return G
```

```python
def adding_name(G):
    
    for each in G.nodes():
        G.node[each]['name'] = random.randint(15,60)
        G.node[each]['type'] = 'person'
```

```python
def add_foci(G):
    n = G.number_of_nodes()
    i = n+1
    foci_nodes = ['Gym', 'eat_out', 'mobieclub', 'karate', 'yoga']
    for j in range(0,5):
        G.add_node(i)
       
        G.node[i]['name'] = foci_nodes[j]
        G.node[i]['type'] = 'foci'
        i = i+1
    
```

```python
#Adding the lables to the graph
def add_labels(G):
    dict1 = {}
    for each in G.nodes():
#         dict1[each].node[each]['name']
        dict1[each] =G.node[each]['name']
    return dict1
```

```python
## Making the size depends on the each node value
def node_size(G):
    array  = []
    for each in G.nodes():
        if G.node[each]['type'] == 'person':
            array.append(G.node[each]['name'])
        else:
            array.append(100)
    return array
```

```python
# Visulization
def visual(G, labels, nsize, colors):
    nx.draw(G,  labels =labels, node_size = nsize, node_color=colors)
    plt.show()
```

```python
def get_colors(G):
    array1 = []
    for each in G.nodes():
        if G.node[each]['type'] == 'person':
            if G.node[each]['name'] == 15:
                array1.append('orange')
            elif G.node[each]['name'] == 40:
                array1.append('green')
            else:
                array1.append('blue')
        else:
            array1.append('red')
    return array1
```

```python
#Adding the homogenity formula is given by probability = 1/abs(difference in nodes)
def homogenity(G):
    pnodes = get_people_nodes()
    for u in pnodes:
        for v in pnodes:
            if u!=v:
                pr = 1/abs(G.node[u]['name']- G.node[v]['name']+1000)
                r = random.uniform(0,1)
                if pr>r :
                    G.add_edge(u, v)
```

```python
#Calling func for creating the graph object
G = create_graph()
```

```python
# Adding the attribute values to the Graph
adding_name(G)
```

```python
add_foci(G)
```

```python
# plotting the graph i.e visulization with labels as bms values
labels = add_labels(G)
```

```python
nsize = node_size(G)
```

```python
colors = get_colors(G)
```

```python
visual(G, labels, nsize, colors)
```

```python
# Adding the edges to this Foci nodes:
def get_foci_nodes():
    f = []
    for i in G.nodes():
        if G.node[i]['type'] == 'foci':
            f.append(i)
    return f
def get_people_nodes():
    p = []
    for i in G.nodes():
        if G.node[i]['type'] == 'person':
            p.append(i)
    return p
def add_foci_edges():
    foci_nodes = get_foci_nodes()
    people_node = get_people_nodes()
    for i in people_node:
        r = random.choice(foci_nodes)
        G.add_edge(i,r)
```

```python
add_foci_edges()
```

```python
visual(G, labels, nsize, colors)
```

```python
homogenity(G)
```
