#### Since the columns names with 'title_x','title_y' has nothing to do with the prediction hence we will be droping the both the columns

```python
new_data.drop(columns= ['title_x','title_y'], inplace = True)
```

#### Since we have some formula for calculating the ratings for the any movie i,e
The weigtage of the any film is calculated by the given formula are as follows, as per IMBD

<img src = 'pavan.png'/>

```python
new_data[['vote_average','vote_count']].sample(3)
```

```python
R = new_data['vote_average']
v = new_data['vote_count']
m = new_data['vote_count'].quantile(0.70)
C =new_data['vote_average'].mean() 
```

```python
new_data['weightage'] = ((R*v)+(C*m))/(v+m)
```

```python
new_data.weightage.sort_values(ascending= False)[:10]
```
