```python
import pandas as pd
```

```python
train = pd.read_csv('cust_segtrain.csv')
```

```python
train.drop(columns = 'ID', inplace = True)
```

```python
train = train.dropna()
```

```python
cat_col = train.select_dtypes(include= 'O')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for col in cat_col.columns:
    le = LabelEncoder()
    train[col] = le.fit_transform(cat_col[col])
```

```python
feature = train.drop(columns = 'Segmentation')
target  =  train.Segmentation
```

```python
from sklearn.model_selection import train_test_split
```

```python
trainX , testX, trainY, testY = train_test_split(feature, target)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
lr = LogisticRegression()
```

```python
lr.fit(trainX, trainY)
```

```python
lr.score(testX, testY)
```

```python
re=  RandomForestClassifier(n_estimators= 100)
```

```python
re.fit(trainX, trainY)
```

```python
re.score(testX, testY)
```
